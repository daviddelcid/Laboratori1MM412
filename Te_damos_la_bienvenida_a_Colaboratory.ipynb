matplotlib.pyplot
Se utiliza para crear gr√°ficos y visualizar datos.


```
import matplotlib.pyplot as plt

x=np.linspace(-10,10,100)   # Generar 100 puntos de -10 a 10
y=x**2  # funcion cuadratica
plt.plot(x,y)
plt.title("grafico de f(x)= x^2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```
